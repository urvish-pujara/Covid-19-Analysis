```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar plots for categorical variables


def plot_bar(data, variable):
    plt.figure(figsize=(8, 6))
    sns.countplot(data=data, x=variable, hue='cov_n')
    plt.title(f'COVID-19 Occurrence by {variable}')
    plt.xlabel(variable)
    plt.ylabel('Count')
    plt.legend(title='COVID-19')
    plt.show()

# Stacked bar plots


def plot_stacked_bar(data, variable):
    ctab = pd.crosstab(data[variable], data['cov_n'], normalize='index')
    ctab.plot(kind='bar', stacked=True, figsize=(8, 6))
    plt.title(f'Stacked Bar Plot of COVID-19 Occurrence by {variable}')
    plt.xlabel(variable)
    plt.ylabel('Proportion')
    plt.legend(title='COVID-19')
    plt.show()

# Heatmap of contingency table


def plot_heatmap(data):
    ctab = pd.crosstab(data['age_n'], [data['sex_n'],
                       data['mar_n']], margins=True)
    plt.figure(figsize=(10, 8))
    sns.heatmap(ctab, annot=True, cmap='YlGnBu', fmt='d')
    plt.title('Heatmap of COVID-19 Occurrence by Age Group, Sex, and Marital Status')
    plt.xlabel('Sex and Marital Status')
    plt.ylabel('Age Group')
    plt.show()

# Box plot for continuous variable


def plot_box(data, variable):
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=data, x='cov_n', y=variable)
    plt.title(f'Box Plot of COVID-19 Occurrence by {variable}')
    plt.xlabel('COVID-19')
    plt.ylabel(variable)
    plt.show()

# Violin plot for continuous variable


def plot_violin(data, variable):
    plt.figure(figsize=(8, 6))
    sns.violinplot(data=data, x='cov_n', y=variable)
    plt.title(f'Violin Plot of COVID-19 Occurrence by {variable}')
    plt.xlabel('COVID-19')
    plt.ylabel(variable)
    plt.show()


# Example usage
plot_bar(data, 'age_n')
plot_stacked_bar(data, 'sex_n')
plot_heatmap(data)
plot_box(data, 'age_n')
plot_violin(data, 'age_n')
```

```python
from scipy.stats import f_oneway
import scipy.stats as stats


# step1 : check the normality of the data
def check_normality(data, factor_column, dependent_column):
    groups = data.groupby(factor_column)[dependent_column]
    normality_results = {}
    for name,group in groups: 
        shapiro_stat, shapiro_p = stats.shapiro(group)
        normality_results[name] = {'Shapiro-Statistic': shapiro_stat, 'p-value': shapiro_p}
    return normality_results

# Step 2: Check Homogeneity of Variances
def check_homogeneity_variances(data, factor_column, dependent_column):
    groups = data.groupby(factor_column)[dependent_column]
    levene_results = stats.levene(*[group for name, group in groups])
    return levene_results


# step3: choose the appropriate test
def choose_test(data, factor_column, dependent_column):
    normality_results = check_normality(data, factor_column, dependent_column)
    levene_results = check_homogeneity_variances(
        data, factor_column, dependent_column)


    if all('p-value' in result and result['p-value'] > 0.05 for result in normality_results.values()) and levene_results.pvalue > 0.05:
        print('The data meets the assumptions of ANOVA')
        test = 'ANOVA'
    else:
        print('The data is not normally distributed or the variances are not homogeneous. Using Kruskal-Wallis Test.')
        test = 'Kruskal-Wallis'
    
    return test

# Step 4: Conduct ANOVA or Kruskal-Wallis Test
def conduct_test(data, factor_column, dependent_column, test):
    groups = [group for name, group in data.groupby(factor_column)[
        dependent_column]]

    if test == 'ANOVA':
        f_statistic, p_value = stats.f_oneway(*groups)
        return f_statistic, None, p_value
    else:
        kruskal_stat, p_value = stats.kruskal(*groups)
        return None, kruskal_stat, p_value


# Step 5: Post-hoc Tests (if applicable)
# Post-hoc tests like Tukey HSD or Bonferroni correction can be conducted using additional libraries like statsmodels.

# Step 6: Report Effect Size
def calculate_effect_size(data, factor_column, dependent_column, test, f_statistic=None, kruskal_stat=None):
    if test == 'ANOVA':
        num_groups = len(data[factor_column].unique())
        num_total = len(data)
        eta_squared = f_statistic / \
            (f_statistic + (num_total - num_groups) / (num_groups - 1))
        effect_size = eta_squared
    else:
        num_total = len(data)
        epsilon_squared = kruskal_stat / (num_total - 1)
        effect_size = epsilon_squared

    return effect_size

# Step 7: Plot Analyzed Data


def plot_data(data, factor_column, dependent_column):
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=factor_column, y=dependent_column, data=data)
    plt.title('Boxplot of ' + dependent_column + ' across ' + factor_column)
    plt.xlabel(factor_column)
    plt.ylabel(dependent_column)
    plt.show()

# Step 8: Interpretation and Reporting


def interpret_results(f_statistic, kruskal_stat, p_value, effect_size, test):
    print('Test:', test)
    print('p-value:', p_value)
    if test == 'ANOVA':
        print('F-statistic:', f_statistic)
        print('Effect size (eta-squared):', effect_size)
    else:
        print('Kruskal-Wallis statistic:', kruskal_stat)
        print('Effect size (epsilon-squared):', effect_size)


# Conducting the Analysis
# Select the columns for the analysis
def performAnalysis(data, factor_column, dependent_columnn):
    normality_results = check_normality(data, factor_column, dependent_columnn)
    levene_results = check_homogeneity_variances(
        data, factor_column, dependent_columnn)
    print('Normality Results:', normality_results)
    print('Levene Results:', levene_results)
    test = choose_test(data, factor_column, dependent_columnn)
    f_statistic, kruskal_stat, p_value = conduct_test(
        data, factor_column, dependent_columnn, test)
    effect_size = calculate_effect_size(
        data, factor_column, dependent_columnn, test, f_statistic, kruskal_stat)
    plot_data(data, factor_column, dependent_columnn)
    interpret_results(f_statistic, kruskal_stat, p_value, effect_size, test)


performAnalysis(data, 'srh_n', 'cov_n')
```
