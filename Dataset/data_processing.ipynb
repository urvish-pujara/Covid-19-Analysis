```python
import pandas as pd 
data = pd.read_csv('Data_COVID19_Engl_sh.csv')
data.head(20)
```

```python
# Age
def map_age(age):
    # if age contains space or is empty, return 7
    age = age.strip()
    if(len(age) > 2):
        return 7
    age = int(age)
    if age >=0 and age <= 65:
        if age <= 20:
            return 1
        elif age <= 34:
            return 2
        elif age <= 44:
            return 3
        elif age <= 54:
            return 4
        elif age <= 64:
            return 5
        elif age >= 65:
            return 6
    else:
        return 7
    
data['age_n'] = data['age'].map(map_age)

# martial status 
data['mar_n'] = data['mar'].map({'Single': 1, 'Married': 2, 'Divorced': 3, 'Other': 4})

# sex 
# Gender
data['sex_n'] = data['sex'].map({'Man': 1, 'Female': 2, 'Other': 3})

# Children's Age
def map_chi_n(chi):
    if chi == "I have no children.":
        return 1
    elif chi == "I am pregnant.":
        return 2
    elif chi == "I am pregnant I have a child/children in the 0-1 age range.":
        return 23
    elif chi == "I am pregnant I have a child/children in the 2-6 age range.":
        return 24
    elif chi == "I am pregnant I have a child/children in the 6-13 age range.":
        return 25
    elif chi == "I have a child/children in the 0-1 age range.":
        return 3
    elif chi == "I have a child/children in the 0-1 age range I have a child/children in the 2-6 age range.":
        return 34
    elif chi == "I have a child/children in the 0-1 age range I have a child/children in the 2-6 age range I have a child/children in the 6-13 age range.":
        return 345
    elif chi == "I have a child/children in the 0-1 age range I have a child/children in the 6-13 age range.":
        return 35
    elif chi == "I have a child/children in the 0-1 age range I have a child/children aged 19 and over.":
        return 37
    elif chi == "I have a child/children in the 2-6 age range.":
        return 4
    elif chi == "I have a child/children in the 2-6 age range I have no children.":
        return 41
    elif chi == "I have a child/children in the 2-6 age range I have a child/children in the 6-13 age range.":
        return 45
    elif chi == "I have a child/children in the 2-6 age range I have a child/children in the 6-13 age range I have a child/children aged 19 and over.":
        return 457
    elif chi == "I have a child/children in the 2-6 age range I have a child/children in the 13-18 age range.":
        return 46
    elif chi == "I have a child/children in the 2-6 age range I have a child/children in the 13-18 age range I have a child/children aged 19 and over.":
        return 467
    elif chi == "I have a child/children in the 2-6 age range I have a child/children aged 19 and over.":
        return 47
    elif chi == "I have a child/children in the 6-13 age range.":
        return 5
    elif chi == "I have a child/children in the 6-13 age range I have a child/children in the 13-18 age range.":
        return 56
    elif chi == "I have a child/children in the 6-13 age range I have a child/children aged 19 and over.":
        return 57
    elif chi == "I have a child/children in the 6-13 age range I have a child/children in the 13-18 age range I have a child/children aged 19 and over.":
        return 567
    elif chi == "I have a child/children in the 13-18 age range.":
        return 6
    elif chi == "I have a child/children in the 13-18 age range I have a child/children aged 19 and over.":
        return 67
    elif chi == "I have a child/children aged 19 and over.":
        return 7
    elif chi == "I am pregnant I have a child/children in the 0-1 age range I have a child/children in the 2-6 age range I have a child/children in the 6-13 age range I have a child/children in the 13-18 age range I have a child/children aged 19 and over I have no children.":
        return 2345671
    else:
        return 0

# Children's Age
data['chi_n'] = data['chi'].map(map_chi_n)

# County
data['ucbs'] = data['cou'].apply(lambda x: 1 if x in ["ANKARA", "ISTANBUL", "IZMIR"] else 2)
data['ist'] = data['cou'].apply(lambda x: 1 if x == "ISTANBUL" else 2)

# Urban / Rural
data['urb_n'] = data['urb'].map({'Rural': 1, 'Urban': 2})

# Education
edu_map = {'None': 0, 'Primary school': 1, 'Middle School': 2, 'High school': 3,
           'Two-year degree or equivalent': 4, 'Graduate': 5,
           "Postgraduate Education (Master's, Doctorate)": 6}
data['edu_n'] = data['edu'].map(edu_map)


# Education - Mother
data['med_n'] = data['med'].map(edu_map)

# Current Working Situation
data['wor_n'] = data['wor'].map({'Unemployed': 1, 'Unemployed after the covid-19 crisis': 2,
                                 'On unpaid leave after the Covid-19 crisis': 3,
                                 'Unpaid family worker (Doing housework etc)': 4,
                                 'Retired': 5, 'Student': 6, 'Part-time employee': 7,
                                 'Full-time employee': 8, 'Self-employed': 9})

# Sector
data['sec_n'] = data['sec'].map({'Agriculture': 1, 'Construction': 2, 'Education': 3,
                                 'Health': 4, 'Industry': 5, 'Service': 6, 'Other': 7})

# Occupation
def map_occupation(occ):
    if occ == "Executives (Senior Officials and Legislators, Administrative and Commercial Executives, etc)":
        return 1
    elif occ == "Professionals (Science, Engineering, Health, Technology, Law, Social and Cultural Professionals, Teachers and Faculty Members, Business and Management Specialist)":
        return 2
    elif occ == "Technicians, and auxiliary professional groups (Science, Engineering, Business and Management Assistant Specialist, Health and Legal Specialists, etc)":
        return 3
    elif occ == "Staff working in office services (Clerks, Customer Service Officers, Digital and Material Registrars, etc)":
        return 4
    elif occ == "Service and sales staff (Personal Service Worker, Sales and Personal Care Worker, Protection Services Worker, etc)":
        return 5
    elif occ == "Skilled agricultural, forestry and aquaculture workers":
        return 6
    elif occ == "Craftsmen and related workers (Construction and Related Workers, Metal, Machinery and Related Trade Workers, Handicrafts and Printing Workers, Electrical and Electronic Trade Workers, etc)":
        return 7
    elif occ == "Plant and machine operators and assemblers":
        return 8
    elif occ == "Non-qualified occupations (Cleaners and Helpers, Agricultural, Forestry and Fisheries Workers, Mining, Construction, Manufacturing and Transport Workers, Food Preparation Assistants, etc)":
        return 9
    elif occ == "Occupations related to the armed forces (Ranked Armed Forces Personnel, Police Force Members, Other Law Enforcement Personnel)":
        return 0
    else:
        return 9999

data['occ_n'] = data['occ'].map(map_occupation)

# parental occupation during childhood
data['foc_n'] = data['foc'].map(map_occupation)

# Home office working conditions 
data['hom_n'] = data['hom'].map({'Yes': 1, 'No': 2, 'Partially': 3})

# Income
def map_income(iin):
    if iin == "0 - 1500":
        return 1
    elif iin == "1500 - 3000":
        return 2
    elif iin == "3000 - 6000":
        return 3
    elif iin == "6000 - 12000":
        return 4
    elif iin == "12000 - 24000":
        return 5
    elif iin == "24000 and above":
        return 6
    else:
        return 0

data['iin_n'] = data['iin'].map(map_income)

# previous year income
data['hin_n'] = data['hin'].map(map_income)

# Self-rated Health
data['srh_n'] = data['srh'].map({'Very good': 5, 'Good': 4, 'Neither good nor bad': 3,
                                 'Bad': 2, 'Very bad': 1}).fillna(0).astype(int)

# Do you stock for one or more items?
data['sto_n'] = data['sto'].map({'Yes': 1, 'No': 0})

# Diagnosed Chronic Illness
data['chr_n'] = data['chr'].map({'Yes': 1, 'No': 0})

# Diagnosed with COVID-19
data['cov_n'] = data['cov'].map({'Yes': 1, 'No': 0})

# #[Considering your online spending habits now compared to pre-COVID-19, which statement is more appropriate for you?] (int)
def map_spend(spe):
    if spe == "Increased a lot":
        return 1
    elif spe == "Increased":
        return 2
    elif spe == "Same":
        return 3
    elif spe == "Decreased":
        return 4
    elif spe == "Decreased a lot":
        return 5
    else:
        return 0

data['int_n'] = data['int'].map(map_spend)

#[Do you think that COVID-19 will have consequences that threaten your personal security?] (dan)

def map_dan(dan):
    if dan == "I definitely think so":
        return 1
    elif dan == "Yes, I think":
        return 2
    elif dan == "I don't know; I'm uncertain":
        return 3
    elif dan == "No, I don't think":
        return 4
    elif dan == "I definitely don't think so":
        return 5
    else:
        return 0
    
data['dan_n'] = data['dan'].map(map_dan)

#[Do you think the state is protecting your individual and community security (sufficiently) during the COVID-19 process?] (gov)
data['gov_n'] = data['gov'].map(map_dan)

#[Which statement is more appropriate for you now when you consider your personal hygiene behavior compared to before COVID-19?] (hyg)
data['hyg_n'] = data['hyg'].map(map_spend)

#[Which statement about your debit/credit card usage is more appropriate for you now compared to pre-COVID-19?] (cre)
data['cre_n'] = data['cre'].map(map_spend)


#The State-Trait Anxiety Inventory 
sur_variables = ["sur1", "sur2", "sur3", "sur4", "sur5", "sur6", "sur7", "sur8", "sur9", "sur10", 
                 "sur11", "sur12", "sur13", "sur14", "sur15", "sur16", "sur17", "sur18", "sur19", "sur20"]

for var in sur_variables:
    data[var + "_n"] = data[var].map({"None": 1, "Low": 2, "Moderate": 3, "High": 4}).fillna(0).astype(int)

sur_n_variables = [var + "_n" for var in sur_variables]
data['surA'] = data[sur_n_variables].sum(axis=1)
surA_summary = data['surA'].describe()

# Create 'surAD' column based on a condition
data['surAD'] = (data['surA'] < 47.94).astype(int)

#The Measurement of Distress Intolerance
dis_variables = ["sur21", "sur22", "sur23", "sur24", "sur25", "sur26", "sur27", "sur28", "sur29", "sur30"]
for var in dis_variables:
    data[var + "_n"] = data[var].map({"Extremely": 4, "Quite a bit": 3, "Moderately": 2, "A little": 1, "Very slighlty or not at all" : 0 }).fillna(0).astype(int)

dis_n_variables = [var + "_n" for var in dis_variables]
data['surB'] = data[dis_n_variables].sum(axis=1)
surB_summary = data['surB'].describe()
data['surBD'] = (data['surB'] < 16.26).astype(int)

# remove first 53 columns
# data = data.iloc[:, 61:]
# save the data
data.to_csv('Data_COVID19_Engl_sh_processed.csv', index=False)

```
